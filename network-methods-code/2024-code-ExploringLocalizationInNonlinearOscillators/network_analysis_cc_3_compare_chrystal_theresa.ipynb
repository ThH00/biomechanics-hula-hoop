```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
chrystal_data_wxy = np.load("chrystal_rotation_results/data_cc_w_xy.npy")
chrystal_data_psidot = np.load("chrystal_rotation_results/data_cc_psidot.npy")
theresa_data_wxy = np.load("theresa_rotation_results/data_th_w_xy.npy")
theresa_data_psidot = np.load("theresa_rotation_results/data_th_psidot.npy")
```

```python
for hoop_quantity in ['w_xy','psidot']:
    print(hoop_quantity)

    chrystal_data = np.load(f"chrystal_rotation_results/data_cc_{hoop_quantity}.npy")
    theresa_data = np.load(f"theresa_rotation_results/data_th_{hoop_quantity}.npy")
    print(f"{chrystal_data.shape=}")
    print(f"{theresa_data.shape=}")
    assert np.allclose(chrystal_data[:,:,1].flatten(),theresa_data[:,1::2].flatten())
    fig,ax = plt.subplots(figsize=(7,3))
    ax.plot(chrystal_data[:,:,1])
    ax.set_title(f"{hoop_quantity} chrystal")
    fig,ax = plt.subplots(figsize=(7,3))
    ax.plot(theresa_data[:,1::2])
    ax.set_title(f"{hoop_quantity} theresa")

    # for coeff in ['C_xys','C_yxs','T_xys','T_yxs']:
    #     print(coeff)
    #     chrystal_coeff = np.load(f"chrystal_rotation_results/{coeff}_{hoop_quantity}.npy")
    #     theresa_coeff = np.load(f"chrystal_rotation_results/{coeff}_{hoop_quantity}.npy")
    #     assert np.allclose(chrystal_coeff,theresa_coeff)
    #     fig,ax = plt.subplots(figsize=(4,4))
    #     ax.imshow(chrystal_coeff)
    #     fig,ax = plt.subplots(figsize=(4,4))
    #     ax.imshow(theresa_coeff)

for coeff in ['C_xys','C_yxs','T_xys','T_yxs']:
    print(coeff)
    chrystal_coeff = np.load(f"chrystal_rotation_results/{coeff}.npy")
    theresa_coeff = np.load(f"theresa_rotation_results/{coeff}.npy")
    assert np.allclose(chrystal_coeff,theresa_coeff)
    fig,ax = plt.subplots(figsize=(4,4))
    ax.imshow(chrystal_coeff)
    ax.set_title(f"{coeff} chrystal")
    fig,ax = plt.subplots(figsize=(4,4))
    ax.imshow(theresa_coeff)
    ax.set_title(f"{coeff} theresa")

```
