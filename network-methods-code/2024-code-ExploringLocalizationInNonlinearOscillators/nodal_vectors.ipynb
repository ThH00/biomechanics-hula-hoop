# Nodal Vectors

## Angular velocities with time
12 nodes: each sensor gets 3 nodes $\in \mathbb{R}^{2}$

For femur, tibia, cuneiform, and hoop:
$$
\begin{bmatrix}
\omega_{x} \\ t
\end{bmatrix}
,\quad{}
\begin{bmatrix}
\omega_{y} \\ t
\end{bmatrix}
,\quad{}
\begin{bmatrix}
\omega_{z} \\ t
\end{bmatrix}
$$


## Euler phase space with time
12 nodes: each sensor gets 3 nodes $\in \mathbb{R}^{3}$

For femur, tibia, cuneiform, and hoop:
$$
\begin{bmatrix}
\phi \\ \dot{\phi} \\ t
\end{bmatrix}
,\quad{}
\begin{bmatrix}
\theta \\ \dot{\theta} \\ t
\end{bmatrix}
,\quad{}
\begin{bmatrix}
\psi \\ \dot{\psi} \\ t
\end{bmatrix}
$$


## Linear phase space with time
12 nodes: each sensor gets 3 nodes $\in \mathbb{R}^{3}$

For femur, tibia, cuneiform, and hoop:
$$
\begin{bmatrix}
d_{x} \\ v_{x} \\ t
\end{bmatrix}
,\quad{}
\begin{bmatrix}
d_{y} \\ v_{y} \\ t
\end{bmatrix}
,\quad{}
\begin{bmatrix}
d_{z} \\ v_{z} \\ t
\end{bmatrix}
$$


## 3D Angular velocities with time

4 nodes: each sensor gets 1 node $\in \mathbb{R}^{4}$

For femur, tibia, cuneiform, and hoop:
$$
\begin{bmatrix}
\omega_{x} \\ \omega_{y} \\ \omega_{z} \\ t
\end{bmatrix}
$$

## 3D Euler phase space with time

4: each sensor gets 1 node $\in \mathbb{R}^{7}$

For femur, tibia, cuneiform, and hoop:
$$
\begin{bmatrix}
\phi \\ \dot{\phi} \\ \theta \\ \dot{\theta} \\ \psi \\ \dot{\psi} \\ t
\end{bmatrix}
$$

## 3D Linear phase space with time

4: each sensor gets 1 node $\in \mathbb{R}^{7}$

For femur, tibia, cuneiform, and hoop:
$$
\begin{bmatrix}
d_{x} \\ v_{x} \\ d_{y} \\ v_{y}  \\ d_{z} \\ v_{z}  \\ t
\end{bmatrix}
$$
